# Ordinal Encoding 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../customer.csv")
df.head()
```

```python
df["age"].nunique()
```

### gender is nominal data and is one hot encoded

```python
df = df.iloc[:, 2:]
```

```python
df.head()
```

```python
input_cols = df.columns.to_list()[0:2]
target_cols = df.columns.to_list()[2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest,Ytrain, Ytest = train_test_split(df[input_cols], df[target_cols], test_size = 0.2)
```

```python
Xtrain.head(), Xtest.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oencoder = OrdinalEncoder(categories=[ ["School", "UG", "PG"], ["Poor", "Average", "Good"]])
```

```python
oencoder.fit(Xtrain)
Xtrain = oencoder.transform(Xtrain)
Xtest = oencoder.transform(Xtest)
```

```python
Xtrain = pd.DataFrame(Xtrain, columns=input_cols)
Xtest = pd.DataFrame(Xtest, columns=input_cols)
```

```python
Xtrain.head(), Xtest.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(Ytrain)
Ytrain = le.transform(Ytrain)
Ytest = le.transform(Ytest)
```
